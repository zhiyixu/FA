```python
import pandas as pd
import numpy as np
from pylab import mpl, plt

plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'

%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\Jason\Desktop\Jason\datas\etf\etf.csv",
header=None, parse_dates=[0])

data.columns = ["date", "UnitPrice", "TotalPrice"]
data.set_index("date", inplace=True)
```

```python
data["UnitPrice"].plot(figsize=(16,10))
```

```python
data['SMA_short'] = data["UnitPrice"].rolling(window=7).mean()
data['SMA_long'] = data["UnitPrice"].rolling(window=30).mean()
data['positions'] = np.where(data['SMA_short'] > data['SMA_long'], 1, -1)
ax = data[["UnitPrice", 'SMA_short', 'SMA_long']].plot(figsize=(10, 6), 
) # notice the secondary_y param secondary_y='positions'
ax.get_legend().set_bbox_to_anchor((.25, .85))
```

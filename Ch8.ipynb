```python
data.resample('1w', label='right').last().head() # fre with one week .
# if resample, padas keep the left point as default, here we use label with right
```

```python
sym = 'AAPL.O'
data = pd.DataFrame(data[sym]).dropna()
data.tail()
```

```python
windows = 20

data['min'] = data[sym].rolling(window=windows).min()
data['mean'] = data[sym].rolling(window=windows).mean()
data['std'] = data[sym].rolling(window=windows).std()
data['median'] = data[sym].rolling(window=windows).median()
data['max'] = data[sym].rolling(window=windows).max()
data['ewma'] = data[sym].ewm(halflife=.5, min_periods=windows).mean()
```

```python
data.dropna().head()
```

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8)
data[sym].iloc[-200:].plot(ax=ax, lw=2.0) # the ax means polt them to the same figure. 
```
